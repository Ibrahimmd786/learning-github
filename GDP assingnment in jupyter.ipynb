```python
all_files=glob.glob(path+"\\*csv")
```

```python
li=[]

for filename in all_files:
    df=pd.read_csv(filename,index_col=None,header=0,encoding='unicode_escape')
    df=df[['Item','2014-15']]
    df=df.T
    df.columns=df.iloc[0]
    df=df[1:]
    df['state']=filename.split('\\')[-1].split('-')[1]
    li.append(df)
frame=pd.concat(li,axis=0,ignore_index=True,sort=True)  
frame
```

```python
frame1=frame.set_index('state')
frame1
```

```python
frame1
df1=frame1[['Primary', 'Secondary','Tertiary','Gross State Domestic Product']] 
df1
df1.loc[:,'primarycontribution'] = df1['Primary']/df1['Gross State Domestic Product']
df1.loc[:,'Secondarycontribution'] =df1['Secondary']/df1['Gross State Domestic Product']
df1.loc[:,'Tertiarycontribution']=df1['Tertiary']/df1['Gross State Domestic Product']
df1
```

```python
#ploting the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP for all the states

df1['Total']=df1['primarycontribution']+df1['Secondarycontribution']+df1['Tertiarycontribution']
df1.sort_values(by='Total',inplace=True,ascending=False)
df1.reset_index(inplace=True)
x=df1[['state','primarycontribution','Secondarycontribution','Tertiarycontribution']]
y=x.set_index('state')
y.plot.bar(stacked=True,width=0.90,figsize=(15,5),title= '%contribution of primary,secondary and tertiary sectors to total GDP for all the states.')
plot.ylabel( "contribution*100")
plot.show
```

```python
#creating a data frame with sub-sectors and other values like Total gsva, Taxes,subsidies,GSDP,percapita 
frame4=frame1[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services',
               'Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting',
               'Financial services','Real estate, ownership of dwelling & professional services','Public administration',
               'Other services','TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products','Gross State Domestic Product',
               'Per Capita GSDP (Rs.)']]
frame4
frame5=frame4.sort_values(by="Per Capita GSDP (Rs.)",axis=0,ascending=False)

frame5
frame5.reset_index(inplace=True)
frame5

```

```python
#calculated the quantile value for percapita column for applying the condition to states
frame6=frame5['Per Capita GSDP (Rs.)'].quantile([0.2,0.5,0.85,1])
frame6
```

```python
# qantile category is present in the last row towards right. please refer to the column
conditions= [
    (frame5['Per Capita GSDP (Rs.)']>153246),
    (frame5['Per Capita GSDP (Rs.)']<=153246)&  (frame5['Per Capita GSDP (Rs.)']>104977),
    (frame5['Per Capita GSDP (Rs.)']<=104977) &    (frame5['Per Capita GSDP (Rs.)']>65187),
    (frame5['Per Capita GSDP (Rs.)']<=65187)]
choices=['C1','C2','C3','C4']
frame5['quantile category']=np.select(conditions,choices,default='')
frame5
```

```python
# removed all the columns which are not usefull for the Categorisation purpose.
frame6=frame5[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services',
               'Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting',
               'Financial services','Real estate, ownership of dwelling & professional services','Public administration',
               'Other services','Gross State Domestic Product','quantile category']]
frame6
```

```python
frame6=frame6.set_index('quantile category')
frame6
```

```python
#categorized the C1 group from the dataframe  
frame_df=frame6.loc['C1']
frame_df
framec1=frame_df.sum(axis = 0, skipna = True)

f1=pd.DataFrame(framec1)
f1.columns=['f1_sum']
f1.loc[:,'f1(%)']=f1['f1_sum']/101930677.0
final_1=f1.drop('Gross State Domestic Product')

fina_l1=final_1.sort_values(by=['f1(%)'],axis=0,ascending=False)
fina_l1

#finding out cumsum for knowing 80% contribution
final1=fina_l1.cumsum()
final1

#to slice the sectors which are contributing 80%
f_inal1= final1[0:7]
f_inal1


```

```python
 #using the GSDP values and slicing it to 80% for the ploting purpose
fig_final1=fina_l1[0:7]
fig_final1
```

```python
fig_final1.plot(y='f1(%)',kind='bar',figsize=(15,5),title= '80% of the GSDP contribution of sub-sectors in C1 category.')
```
